```python
ladder = Ladder()
```

```python
l1, n11, n12 = ladder.add_cycle("1", "11", "12")
l2, n21, n22, n23, n24, n25 = ladder.add_cycle("2", "21", "22", "23", "24", "25")
l3, n31, n32, n33, n34, n35 = ladder.add_cycle("3", "31", "32", "33", "34", "35")
l4, n41, n42 = ladder.add_cycle("4", "41", "42")
```

```python
n11.hori_connect(n24)
n12.hori_connect(n22)
n21.hori_connect(n35)
n23.hori_connect(n33)
n25.hori_connect(n31)
n41.hori_connect(n34)
n42.hori_connect(n32)
```

```python
ladder.fix()
```

```python
ladder.dynamism(10)
```

```python
drawer = LadderDrawer(ladder2)
drawer.draw_ghost_leg_ellipse(check_point=True)
```
